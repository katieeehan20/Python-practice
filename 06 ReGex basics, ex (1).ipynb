ReGex basics



```python
import re
text_to_search='''
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890
Ha Haha

Metacharacters(Need to be escaped):  #special characters
. ^ $ * + ? { } [ ] \ | ( )

coreyms.com   #URL
321-555-4321  #phone number
123.555.1234
Mr.Schafer    #name
Mr Smith
Ms Davis
Mrs.Robinson
Mr. T
'''

sentence="Start a sentence and then bring it to an end"
```

```python
print('\tTab')
```

```python
print(r'abc')  # There shouldn't be space between r and 'abc'
```

```python
pattern=re.compile(r'abc')
```

```python
matches=pattern.finditer(text_to_search)
```

```python
for match in matches:
    print(match)       # The result means we found 1 match of 'abc' and found it in our alphabet from the index 1 to 4
```

```python
print(text_to_search[1:4])  # string slicing functionality
```

```python
pattern=re.compile(r'cba')  # There are no matches for cba
```

```python
pattern=re.compile(r'.')     # returns almost all values in the multi-line string. 
                             # . is a special character in ReGex
                             # . matches any character without new line

matches=pattern.finditer(text_to_search)
for match in matches:
    print(match) 
```

```python
# if you want to just search for a period, you have to escape the . character. 
# you can escape the character by placing a backslash...:


pattern=re.compile(r'\.')  #backslash is placed
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# to find a URL, because it contains a dot, do the following:

pattern=re.compile(r'coreyms\.com')
matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# [1-5] matches any digits between 1 and 5

pattern=re.compile(r'[1-5]')

matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# [a-z] matches any digits between 1 and 5

pattern=re.compile(r'[a-z]')

matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python

pattern=re.compile(r'[a-zA-Z]')

matches=pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
#[^] returns anything except a certain character within [], in this case [^b] excludes bat

new_line="""
cat
mat
pat
bat
rat
"""


pattern=re.compile(r'[^b]at')  

matches=pattern.finditer(new_line)
for match in matches:
    print(match)
```
