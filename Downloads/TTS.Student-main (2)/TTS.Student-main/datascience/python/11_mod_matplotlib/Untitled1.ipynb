```python
# setting up the notebook for plotting and importing the packages we'll use 

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
# for all matplotlib plots, we start by creating a figure and axes

fig= plt.figure()
ax=plt.axes()
```

```python
fig= plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x));
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
# to change color

plt.plot(x, np.sin(x), color='blue')  
plt.plot(x, np.sin(x-1), color='g')        # short color code
plt.plot(x, np.sin(x-2), color='0.5')      # greyscale
plt.plot(x, np.sin(x-3), color='#FFDD44')  # hexcode
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))  # RGB tuple
plt.plot(x, np.sin(x - 5), color='chartreuse')   # HTML color names
```

```python
# to change line style

plt.plot(x, x, linestyle='solid', color= 'magenta')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dotted')
plt.plot(x, x+3, linestyle='dashdot')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle=':')
plt.plot(x, x+7, linestyle='-.')
```

```python
plt.plot(x, x, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```
